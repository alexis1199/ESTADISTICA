## Introduccion:
El presente conjunto de datos contiene información médica y demográfica de pacientes, con el objetivo de analizar factores asociados a la diabetes y su posible relación con la mortalidad. Cada registro representa a un paciente, incluyendo variables clínicas como el nivel de glucosa, presión arterial, grosor del pliegue cutáneo, niveles de insulina, índice de masa corporal (BMI), así como factores de riesgo hereditario y edad.

## Hipotesis:
Los niveles elevados de glucosa en sangre, junto con un índice de masa corporal (BMI) alto, insulina y antecedentes hereditarios de diabetes, están significativamente asociados con una mayor probabilidad de mortalidad en pacientes con riesgo de diabetes.

## Marco Teorico
La diabetes es una enfermedad metabólica crónica caracterizada por niveles elevados de glucosa en sangre, resultado de defectos en la secreción o acción de la insulina, o ambos. Esta condición representa un importante problema de salud pública a nivel mundial debido a su alta prevalencia y a las múltiples complicaciones asociadas, entre ellas enfermedades cardiovasculares, daño renal, neuropatías y aumento del riesgo de mortalidad.

## Metodologia
Enfoque
Este estudio adopta un enfoque cuantitativo, descriptivo y correlacional, con el objetivo de analizar la relación entre diferentes variables clínicas y demográficas con la mortalidad en pacientes con factores de riesgo asociados a la diabetes mellitus.

Fuente de Datos
Se utilizó un conjunto de datos clínicos que contiene información de pacientes con posible riesgo de diabetes. El dataset incluye variables como:

Pregnancies: número de embarazos

Glucose: nivel de glucosa en sangre

BloodPressure: presión arterial

SkinThickness: grosor del pliegue cutáneo

Insulin: niveles de insulina

BMI: índice de masa corporal

DiabetesPedigreeFunction: antecedentes familiares de diabetes

Age: edad del paciente

Outcome: estado del paciente (vivo o muerto)

## Conclusión
A lo largo de este análisis, pude observar cómo diversas variables clínicas y demográficas influyen en la probabilidad de mortalidad en pacientes con riesgo de diabetes. Me llamó la atención la fuerte relación entre niveles elevados de glucosa, un índice de masa corporal alto y la función hereditaria con el desenlace fatal reflejado en la variable Outcome. Estos factores no solo son indicadores comunes en el diagnóstico de la diabetes, sino que también parecen jugar un papel clave en la progresión y gravedad de la enfermedad.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[['SkinThickness','BloodPressure', 'Glucose', 'Pregnancies', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
Y = df['Insulin']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)

X_train_constante = sm.add_constant(X_train)
X_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, X_train_constante).fit()


Yc = modelo.predict(X_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinacion: {r2: 0.4f}')
```
